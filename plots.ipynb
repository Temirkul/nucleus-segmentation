```python
import numpy as np
import matplotlib.pyplot as plt 
import torch
import os 
```

```python
num_epochs = 50
learning_rate = 1e-4
drop_rates = [0.01, 0.2, 0.5, 0.8, 0.99]

dict_4_False_list = []
dict_4_True_list = []
dict_5_False_list = []
dict_5_True_list = []
dict_5_False_list_transform = []

val_error_4_False_arr = np.zeros(5)
val_error_4_True_arr = np.zeros(5)
val_error_5_False_arr = np.zeros(5)
val_error_5_True_arr = np.zeros(5)
val_error_5_False_arr_transform = np.zeros(5)

train_error_4_False_arr = np.zeros(5)
train_error_4_True_arr = np.zeros(5)
train_error_5_False_arr = np.zeros(5)
train_error_5_True_arr = np.zeros(5)
train_error_5_False_arr_transform = np.zeros(5)

i = 0
for drop_rate in drop_rates:
    path_4_False = f'C:/Users/Temirkul/Desktop/Temirkul/quantori unet (send to laptop)/saves/unet_{num_epochs}epoch_{drop_rate}_{learning_rate}_invert{False}_{4}.pt'
    path_4_True = f'C:/Users/Temirkul/Desktop/Temirkul/quantori unet (send to laptop)/saves/unet_{num_epochs}epoch_{drop_rate}_{learning_rate}_invert{True}_{4}.pt'
    path_5_False = f'C:/Users/Temirkul/Desktop/Temirkul/quantori unet (send to laptop)/saves/unet_{num_epochs}epoch_{drop_rate}_{learning_rate}_invert{False}_{5}.pt'
    path_5_True = f'C:/Users/Temirkul/Desktop/Temirkul/quantori unet (send to laptop)/saves/unet_{num_epochs}epoch_{drop_rate}_{learning_rate}_invert{True}_{5}.pt'
    path_5_False_transform = f'C:/Users/Temirkul/Desktop/Temirkul/quantori unet (send to laptop)/saves/unet_{num_epochs}epoch_{drop_rate}_{learning_rate}_invert{False}_{5}_transform.pt'
    
    dict_4_False_list.append(torch.load(path_4_False, map_location=torch.device('cpu')))
    dict_4_True_list.append(torch.load(path_4_True, map_location=torch.device('cpu')))
    dict_5_False_list.append(torch.load(path_5_False, map_location=torch.device('cpu')))
    dict_5_True_list.append(torch.load(path_5_True, map_location=torch.device('cpu')))
    dict_5_False_list_transform.append(torch.load(path_5_False_transform, map_location=torch.device('cpu')))
    
    val_error_4_False_arr[i] = dict_4_False_list[i]['val_loss_arr'][-1]
    train_error_4_False_arr[i] = dict_4_False_list[i]['train_loss_arr'][-1]
    
    val_error_4_True_arr[i] = dict_4_True_list[i]['val_loss_arr'][-1]
    train_error_4_True_arr[i] = dict_4_True_list[i]['train_loss_arr'][-1]
    
    val_error_5_False_arr[i] = dict_5_False_list[i]['val_loss_arr'][-1]
    train_error_5_False_arr[i] = dict_5_False_list[i]['train_loss_arr'][-1]
    
    val_error_5_True_arr[i] = dict_5_True_list[i]['val_loss_arr'][-1]
    train_error_5_True_arr[i] = dict_5_True_list[i]['train_loss_arr'][-1]
    
    val_error_5_False_arr_transform[i] = dict_5_False_list_transform[i]['val_loss_arr'][-1]
    train_error_5_False_arr_transform[i] = dict_5_False_list_transform[i]['train_loss_arr'][-1]
    
    print(i)
    
    i += 1
```

```python
plt.plot(drop_rates, val_error_4_False_arr, label='depth=4, invert=False, transform=False', linestyle='--', marker='o')
plt.plot(drop_rates, val_error_4_True_arr, label='depth=4, invert=True, transform=False', linestyle='--', marker='o')
plt.plot(drop_rates, val_error_5_False_arr, label='depth=5, invert=False, transform=False', marker='o')
plt.plot(drop_rates, val_error_5_True_arr, label='depth=5, invert=True, transform=False', marker='o')
plt.plot(drop_rates, val_error_5_False_arr_transform, label='depth=5, invert=False, transform=True', marker='o')
plt.xlabel('Drop rates')
plt.ylabel('CE Loss')
plt.title('Validation Error')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(drop_rates, train_error_4_False_arr, label='depth=4, invert=False, transform=False', linestyle='--', marker='o')
plt.plot(drop_rates, train_error_4_True_arr, label='depth=4, invert=True, transform=False', linestyle='--', marker='o')
plt.plot(drop_rates, train_error_5_False_arr, label='depth=5, invert=False, transform=False', marker='o')
plt.plot(drop_rates, train_error_5_True_arr, label='depth=5, invert=True, transform=False', marker='o')
plt.plot(drop_rates, train_error_5_False_arr_transform, label='depth=5, invert=False, transform=True', marker='o')
plt.xlabel('Drop rates')
plt.ylabel('CE Loss')
plt.title('Train Error')
plt.grid()
plt.legend()
plt.show()
```
